# Excel käyttöliittymänä Pythonin skriptien käynnistämiseen

Exceliä voi käyttää käyttöliittymänä python-skriptien käynnistämiseen. Tätä varten pitää asentaa Add-in. Sen voi asentaa Anacondan tai Minicondan komentoriviltä: 
$$\textrm{xlwings addin install}$$ Asennuskomento on suoritettava uudelleen joka kerta kun päivität xlwings-kirjaston uudempaan versioon!

Asennuksen jälkeen Excelin yläreunan valintanauhasta pitäisi löytyä xlwings-välilehti:

Jos Excelin **Developer** (Kehitystyökalut) eivät ole Excelin yläreunan valintanauhassa, niin hae ne käyttöön:

Valitse **File – Options – Customize Ribbon** (Tiedosto – Asetukset – Valintanauhan mukauttaminen) ja valitse **Developer** (Kehitystyökalut).

## Python-skriptin käynnistys Excelistä

Python-skripti eli .py-tiedostossa oleva ohjelmakoodi voidaan käynnistää Excelistä ainakin kahdella seuraavaksi esitettävällä tavalla. Niistä ensimmäinen on yksinkertaisempi ja sopii moneen käytännön tilanteeseen vallan mainiosti. Toinen soveltuu monipuolisesti erilaisiin lähtötilanteisiin mutta vaatii hieman enemmän alkuvalmsteluja. Molemmat tavat on esitetty seuraavassa.

### Yksinkertaisempi tapaus

Tässä ensin kuvattu yksinkertaisempi tapaus toimii sillä edellytyksellä, että seuraavat kaksi ehtoa täyttyvät:

* skripti (.py-tiedosto) on tallennettu samalla nimellä kuin työkirja (.xlsx-tiedosto)
* skripti-tiedostosta löytyy funktio, jonka nimi on **main()**.

Nämä ehdot täyttävä sovellus (**sovellus1**) on rakennettu alla kuvatulla tavalla. Kurssin materiaalipaketista löytyvät alla mainitut samannimiset tiedostot ja hakemisto.

* Ensin luotiin sovellus1-niminen kansio.
* Skripti **sovellus1.py** tallennettiin sovellus1-kansioon.
* Luotiin uusi Excel-tiedosto, johon tehtiin valintalista osakkeen valitsemista varten.
* Siirryttiin Excelin ohjelmointiympäristöön valitsemalla valintanauhan **Developer** (Kehitystyökalut) -välilehdeltä **Visual Basic**.
* Valittiin Excelin ohjelmointiympäristössä **Tools – References – xlwings** (tämä on työkirjakohtainen määritys).
* Siirryttiin takaisin Exceliin (siirtyminen onnistuu esimerkiksi sulkemalla Excelin ohjelmointiympäristön ikkunan).
* Tallennettiin Excel-tiedosto sovellus1-kansioon nimellä **sovellus1.xlsx** (*on tärkeää että skriptillä ja Excel-tiedostolla on sama nimi*).
* Edellisten vaiheiden jälkeen valintanauhan xlwings-välilehden **Run main** -painike suorittaa skriptin sovellus1.py, joka hakee valintalistasta valitun osakkeen historiatiedot ja piirtää Bollingerin nauhat kurssihistoria-taulukkovälilehdelle.

### Monipuolisempi tapaus

Jos jompikumpi yllä mainituista ehdoista ei täyty, niin työkirjaan pitää lisätä VBA-kielinen makro, joka kutsuu Python-skriptiä. Jotta VBA-makro toimisi niin työkirja täytyy tallentaa xlsm-muodossa. Kurssin materiaalipaketin **sovellus2**-kansiossa on tiedostot tällaista tapausta kuvaavaan esimerkkiin, jossa kolme VBA-makroa kutsuvat **sovellus2.py**-tiedoston erinimisiä funktioita. VBA-makrot pystyy lisäämään sovellus2.xlsx-tiedostoon seuraavasti:

* Siirrytään Excelin ohjelmointiympäristöön valitsemalla valintanauhan **Developer** (Kehitystyökalut) -välilehdeltä **Visual Basic**.
* Valitaan Excelin ohjelmointiympäristössä **Tools – References – xlwings** (tämä on työkirjakohtainen määritys).
* Lisätään VBA-moduuli komennolla **Insert - Module**.
* Kirjoitetaan VBA-makrot juuri luotuun moduuliin. Esimerkiksi komento
$$\textrm{RunPython "import sovellus2; sovellus2.osake()"}$$
tuo sovellus2.py-moduulin ja sen jälkeen käynnistää sieltä osake()-nimisen funktion. Tässä ei edellytetä, että skripti ja työkirja ovat samannimisiä (kuten tässä sovellus2).

Excelistä on tavalla tai toisella käynnistettävä VBA-makrot. Tässä voi käyttää valintalistaa ja painikkeita:

* Klikataan hiiren kakkospainikkeella valintalistan päällä ja valitaan Assign Macro (Liitä Makro).
* Valitaan makro, joka halusin liittää valintalistaan.
* Valitaan **Developer** (Kehitystyökalut)-välilehdeltä **Insert - Button** (Lisää - Painike).
* Piirretään painike.
* Klikataan hiiren kakkospainikkeella painikkeen päällä ja valitaan **Assign Macro** (Liitä Makro).
* Valitaan makro, joka halutaan liittää painikkeeseen.
